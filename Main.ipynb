## **Declaração de função de converção do audio e acesso a pasta raiz da base de dados**

```
def get_spectrogram( wav ):
    D = librosa.stft(wav, n_fft=1080, hop_length=160,
                     window='hamming', center=True)
    spect, phase = librosa.magphase( D )
    return spect


a = os.listdir( 'drive/My Drive/Train2' )
print(' ...OK')
```

---
## **Acesso, conversão, salvamento com liberação**

```
for train_path in a:
    i = 0
    b = os.listdir( 'drive/My Drive/Train2' + '/' + train_path  + '/'  )
    print(train_path)
    print('\n')

    for dir_path in b:
        c = os.listdir( 'drive/My Drive/Train2' + '/' + train_path + '/' + dir_path  + '/'  )
        print(dir_path)
        print('\n')
        for file_path in c:
            print(file_path)
            i = i + 1
            sound = AudioSegment.from_ogg( 'drive/My Drive/Train2/' + train_path + '/' + dir_path + '/' + file_path)
            print(sound)
            sound.export( "audioconvertido.wav", format="wav" )
            wav, sr = librosa.load( "audioconvertido.wav", sr=None )
            print( wav.shape, wav.max(), wav.min() )
            ipd.Audio( wav, rate=sr )
            log_spect = np.log( get_spectrogram( wav ) )
            plt.tick_params(top=False, bottom=False, left=False,
                            right=False, labelleft=False, labelbottom=False)
            plt.imshow( log_spect, aspect='auto', origin='hight' )
            plt.savefig( 'drive/My Drive/Train2/PNG' + '/' + dir_path + '/' + train_path + str(i) + ".jpg", dpi = 100, bbox_inches = 'tight')
            print( "drive/My Drive/Train2/PNG" + dir_path + "/" + train_path + str(i) + ".png" )
            del( log_spect )
            del( wav )
            del( sr )
            del( sound )
            print(' ...OK')
            print('\n')

print("\n\n\n Finalizado.")
```

## **Importação das dependências para a construção da rede neural**

```
from numpy import array
from PIL import Image
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K
import numpy as np
from keras.preprocessing import image
import keras
from keras import regularizers, optimizers

print('...OK')
```

---
##Pre-set da base de dados com normalização e pré-definição de rede

```
img_width, img_height = 150, 150
train_data_dir = 'drive/My Drive/Train2/PNG'
validation_data_dir = 'drive/My Drive/Train2/PNG'

nb_train_samples = 848
nb_validation_samples = 114
epochs = 40
batch_size = 32
if K.image_data_format() == 'wa':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)

train_datagen = ImageDataGenerator(rescale=1./255., validation_split=0.25)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    seed=42,
    shuffle=True,
    class_mode='categorical'
)

validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    seed=42,
    shuffle=True,
    class_mode='categorical'
)

print('\n ...OK')
```

---
## **Criação e definição da rede neural**

```
model = Sequential()
model.add(Conv2D(32, (3, 3), padding='same', input_shape=(150, 150, 3)))
model.add(Activation('relu'))
model.add(Conv2D(150, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(150, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(150, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))
model.add(Conv2D(300, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(300, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(5))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(4, activation='softmax'))

model.compile(optimizers.rmsprop(lr=0.0005, decay=1e-6),
              loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()

print(batch_size)
STEP_SIZE_TRAIN = train_generator.n // train_generator.batch_size
STEP_SIZE_VALID = validation_generator.n // validation_generator.batch_size

print('\n ...OK')
```

---
##**Inicio de treino**

```
hist = model.fit_generator(
        generator=train_generator,
        steps_per_epoch=STEP_SIZE_TRAIN,
        validation_data=validation_generator,
        validation_steps=STEP_SIZE_VALID,
        epochs=100
)
```

---
## **Salvamento de pesos e plotagem**

```
print(batch_size)
model.save_weights('first_try.h5')

plt.plot(hist.history['acc'])
plt.plot(hist.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
